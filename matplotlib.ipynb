```python
t=np.linspace(0,5,20)
plt.plot(t,t**2,color='coral',linestyle='--')
plt.plot(t,4*t,'v--',linewidth=2)
```

```python
plt.subplot(1,3,1)
plt.plot(t,t,'c--')

plt.subplot(1,3,2)
plt.plot(t,t**2,'b--')

plt.subplot(1,4,4)
plt.plot(t,t**3,'mo')

```

```python
grades={'john':70,
       'suraj':80,
       'deepak':90,
       'alex':89}
plt.bar(range(len(grades)),list(grades.values()),color='pink')
plt.xticks(range(len(grades)),list(grades.keys()))
plt.xlabel('students')
plt.ylabel('marks')
```
