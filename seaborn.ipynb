```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Pokemon (2).csv',index_col=0,encoding='unicode-escape')
df.head()

```

```python
sns.lmplot(x='Attack' ,y='Defense', data=df)
plt.show()
```

```python
sns.lmplot(x='Attack', y='Defense',data=df,fit_reg=False,hue='Stage')#fit_reg to remove regression line and hue gives different colors to different stages
plt.show()
```

```python
df_copy=df.drop(['Total','Stage','Legendary'],axis=1)
sns.boxplot(data=df_copy)
```

```python
sns.violinplot(data=df_copy)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='Type 1',y='Attack',data=df)
plt.show()
```

```python
corr=df_copy.corr()
sns.heatmap(corr)
```

```python
sns.distplot(df.Attack,color='blue')#distribution plot
```

```python
sns.countplot(x='Type 1',data=df)
plt.xticks(rotation=-45)
```

```python
sns.jointplot(df.Attack,df.Defense,kind='kde',color='lightblue')
```
